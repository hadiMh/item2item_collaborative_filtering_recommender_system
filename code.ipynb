```python
import pandas as pd
import numpy as np
```

```python
# rating.csv from this link: https://www.kaggle.com/datasets/grouplens/movielens-20m-dataset?select=rating.csv
ratings_org = pd.read_csv('rating.csv')
```

```python
ratings = ratings_org[['userId', 'movieId', 'rating']][:1000000]
ratings.head()
```

```python
ratings_pivot = ratings.pivot(index='userId', columns='movieId', values='rating').T.fillna(0)
ratings_pivot
```

```python
# Just to observe that how many movies each user has scored
ratings_pivot.T.apply(lambda x: sum([1 if num > 1 else 0 for num in x]), axis = 1)
```

```python
# How many scores each movie has
ratings_pivot.apply(lambda x: sum([1 if num > 1 else 0 for num in x]), axis = 1)
```

```python
# What are the movies ids that user with user id user_id has not watched yet
def movies_not_watched_by(user_id):
    watched_movies = ratings[ratings['userId']==user_id]['movieId'].tolist()
    # Diffrence all of the movies from the watched ones
    return set(ratings['movieId']).difference(watched_movies)

movies_not_watched_by(1)
```

```python
from sklearn.neighbors import NearestNeighbors

# Find the K nearest neighbors for every movie based on the ratings
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(ratings_pivot.values)
distances, indices = knn.kneighbors(ratings_pivot.values, n_neighbors=3)
```

```python
indices
```

```python
def recommend_movie(user_id):
    # Get the unwatched movie ids
    # We use this list to predict the scores of them for this user
    movies_not_watched = movies_not_watched_by(user_id)

    user_final_similar_movies_rates = []

    # Iterate through the not watched movies to estimate their scores
    for movie_id in movies_not_watched:
        # Get the index of the movie in the KNN Indices list
        movie_knn_index = np.where(ratings_pivot.index.values == movie_id)
        # Find the list of simialr movies ids
        similar_movies_to_this_movie_that_user_liked = list(indices[movie_knn_index][0])
        similar_movies_distances = list(distances[movie_knn_index][0])
        
        if movie_knn_index in similar_movies_to_this_movie_that_user_liked:
            # The index of the movie id itself in the list
            iii = similar_movies_to_this_movie_that_user_liked.index(movie_knn_index)
            # Remove the movie index itself from the list
            similar_movies_to_this_movie_that_user_liked.pop(iii)
            # Remove the movie distance itself from the list
            similar_movies_distances.pop(iii)
        else:
            # ELSE is for the time that the movie vector is all zero.
            # So it finds similar movies as they are all zero
            # So the movie index itself is not in the list
            # So we remove the furthest in the list which is the last one
            similar_movies_to_this_movie_that_user_liked = similar_movies_to_this_movie_that_user_liked[:-1]
            similar_movies_distances = similar_movies_distances[:-1]

        # The similarities of the movies
        movies_similarities = [1-sim for sim in similar_movies_distances]

        all_predicted_scores = []
        for index, sim_value in enumerate(movies_similarities):
            try: # We use try/except because we do not load all the data as it is very big so some movies are not loaded and we may have Key Error
                # Multiply the score of the similar movie that this user gave to it, to the similary of this movie to the actual movie
                predicted_score = ratings_pivot[similar_movies_to_this_movie_that_user_liked[index]][user_id] * sim_value
                # If user has scores the movie, save the predicted value to calculate later
                if predicted_score != 0:
                    all_predicted_scores.append(predicted_score)
            except KeyError:
                pass

        final = sum(all_predicted_scores)/len(all_predicted_scores) if all_predicted_scores else 0
        user_final_similar_movies_rates.append((movie_id, final))

    return user_final_similar_movies_rates
```

```python
# This function gets the user_id and n(default:3)
# And recommends 3 movies for the specified user
def recommend_n_movies_to_user(user_id, n=3):
    movies = recommend_movie(user_id)
    three_movies = list(sorted(movies, key=lambda x: x[1], reverse=True))[:n]

    return three_movies

# for the user user_id=100 it recommends movies: 3058, 4576, 3933
recommend_n_movies_to_user(100)
```
